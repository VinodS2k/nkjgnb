```python
from google.colab import files

uploaded = files.upload()




from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Load the image (use the uploaded filename)
img = Image.open(next(iter(uploaded)))

# Convert the image to grayscale (if needed)
gray_img = img.convert('L')

# Convert the grayscale image to a numpy array
pixel_values = np.array(gray_img)

# Create a histogram
plt.figure(figsize=(10, 5))
plt.hist(pixel_values.ravel(), bins=256, range=[0, 256], color='gray', alpha=0.7)
plt.title('Histogram of Grayscale Image')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.xlim([0, 256])
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
from google.colab import files

uploaded = files.upload()






from PIL import Image, ImageFilter
import matplotlib.pyplot as plt

# Load the image (use the uploaded filename)
img = Image.open(next(iter(uploaded)))

# Define a function to display images
def display_images(images, titles):
    plt.figure(figsize=(15, 10))
    for i, (image, title) in enumerate(zip(images, titles)):
        plt.subplot(2, 3, i + 1)
        plt.imshow(image, cmap='gray')
        plt.title(title)
        plt.axis('off')
    plt.show()

# Apply different filters
blurred = img.filter(ImageFilter.BLUR)
contour = img.filter(ImageFilter.CONTOUR)
edge_enhance = img.filter(ImageFilter.EDGE_ENHANCE)
sharpened = img.filter(ImageFilter.SHARPEN)
detail = img.filter(ImageFilter.DETAIL)

# Prepare images and titles for display
images = [img, blurred, contour, edge_enhance, sharpened, detail]
titles = ['Original', 'Blurred', 'Contour', 'Edge Enhance', 'Sharpened', 'Detail']

# Display the images
display_images(images, titles)
```
